```python
import pandas as pd
```

```python
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
import io
```

```python
 import io
df = pd.read_csv(io.BytesIO(uploaded['attendance_avg.csv']), delimiter='\t') #reading a csv file, fixed key in dictionary, delimiter='\t' if delimited by tabs
```

```python
df.head(7) #to read first 7 elments
```

```python
df.isnull() # to check wheter there is a null value or not represented by True or False
```

```python
df.isnull().sum() # to show the sum of the null value
```

```python
df.describe() # to count the mean, max and min value of the given csv file
```

```python
df.tail(7) # to print last 7 elemnts of the csv file
```

```python
#Separate the comma-separated values into individual columns
df = df['SAP ID,ROLL ID,NAME,Avg'].str.split(',', expand=True)

# Assign appropriate column names
df.columns = ['SAP ID', 'ROLL ID', 'NAME', 'Avg']

# Convert 'Avg' column to numeric type
df['Avg'] = pd.to_numeric(df['Avg'])

# Now you can calculate the mean
df['Avg'].mean()
```

```python
df.min()# To find the minimum value present in the csv file
```

```python
df.max()# To find the maximum value present in the csv file
```

```python
df.sum() #Sum of numerical columns
```
